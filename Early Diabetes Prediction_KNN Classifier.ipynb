### Pre processing 

```python
#Missing values

print(training_data.isnull().sum())
print()
print(testing_data.isnull().sum())
```

```python
#imputing Age with mean value

training_data["Age"].fillna(int(training_data["Age"].mean()),inplace=True)
print(training_data.isnull().sum())

testing_data["Age"].fillna(int(testing_data["Age"].mean()),inplace=True)
print(testing_data.isnull().sum())
```

```python
#Identify format of missing value

for i in training_data.columns:
    print(training_data[i].unique())
```

```python
#Identify format of missing value

for i in testing_data.columns:
    print(testing_data[i].unique())
```

```python
#imputing others with mode value

for value in ['sudden weight loss','muscle stiffness']:
    training_data[value].fillna(training_data[value].mode()[0],inplace=True)
print(training_data.isnull().sum())

for value in ['sudden weight loss','muscle stiffness']:
    testing_data[value].fillna(testing_data[value].mode()[0],inplace=True)
print(testing_data.isnull().sum())
```

```python
#Label encoding

colname=[]
for x in training_data.columns:
    if training_data[x].dtype=='object':
        colname.append(x)
```

```python
from sklearn import preprocessing
le=preprocessing.LabelEncoder()

for x in colname:
    training_data[x]=le.fit_transform(training_data[x]) 
    testing_data[x]=le.fit_transform(testing_data[x]) 
```

```python
le_name_mapping=dict(zip(le.classes_,le.transform(le.classes_)))
print('Feature',x)
print('mapping',le_name_mapping) 
```

```python
# Splitting Data into dep and ind vars

X_train=training_data.values[:,0:-1] 
Y_train=training_data.values[:,-1]
Y_train=Y_train.astype(int)

X_test=testing_data.values[:,0:-1]
Y_test=testing_data.values[:,-1]
Y_test=Y_test.astype(int)
```

```python
#Scaling data

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(X_train)

X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

### KNN Model

```python
# Approach1: K value specification - SQRT of train data Observations

from sklearn.neighbors import KNeighborsClassifier
model_KNN=KNeighborsClassifier(n_neighbors=int(np.sqrt(len(X_train))),metric='euclidean')
model_KNN.fit(X_train,Y_train)
Y_pred=model_KNN.predict(X_test)
print(list(zip(Y_test,Y_pred)))
```

```python
np.sqrt(len(X_train))
```

```python
# Metrics Evaluation

from sklearn.metrics import confusion_matrix,accuracy_score,classification_report

cfm=confusion_matrix(Y_test,Y_pred)
print(cfm)

print()

print("classification report:")
print(classification_report(Y_test,Y_pred))

print()

acc=accuracy_score(Y_test, Y_pred)
print("Accuracy of the model: ",acc)
```

```python
# Approach2: K value specification - trail and error 

from sklearn.metrics import accuracy_score
my_dict={}
for K in range(1,26):
    model_KNN = KNeighborsClassifier(n_neighbors=K,metric="euclidean")
    model_KNN.fit(X_train, Y_train)
    Y_pred = model_KNN.predict(X_test)
    print ("Accuracy is ", accuracy_score(Y_test,Y_pred), "for K-Value:",K)
    my_dict[K]=accuracy_score(Y_test,Y_pred)
```

###  Model2: Logistic Regression as its binary classification. To Compare which Algorithm is better

```python
from sklearn.linear_model import LogisticRegression
classifier=LogisticRegression()
classifier.fit(X_train,Y_train)
Y_pred=classifier.predict(X_test)
print(list(zip(Y_test,Y_pred)))
```

```python
# Metrics Evaluation

from sklearn.metrics import confusion_matrix,accuracy_score,classification_report

cfm=confusion_matrix(Y_test,Y_pred)
print(cfm)

print()

print("classification report:")
print(classification_report(Y_test,Y_pred))

print()

acc=accuracy_score(Y_test, Y_pred)
print("Accuracy of the model: ",acc)
```
