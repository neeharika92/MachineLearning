### Pre Processing

```python
#missing values
data.isnull().sum()
```

```python
#Label Encoding

from sklearn import preprocessing
colname=['diagnosis']
le=preprocessing.LabelEncoder()

for x in colname:
    data[x]=le.fit_transform(data[x]) 
```

```python
#Optional code: To know which label is associated with which category (for printing dictionary)

le_name_mapping=dict(zip(le.classes_,le.transform(le.classes_)))
print('Feature',x)
print('mapping',le_name_mapping) 
```

```python
data.head()
```

```python
#Splitting data into dep and ind vars

X=data.values[:,2:] 
Y=data.values[:,1]
Y=Y.astype(int)
```

```python
#Scaling Data

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(X)
X = scaler.transform(X)
print(X)
```

### KNN Model

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2,random_state=10)
```

```python
#K value - SQRT method

from sklearn.neighbors import KNeighborsClassifier
model_KNN=KNeighborsClassifier(n_neighbors=int(np.sqrt(len(X_train))),metric='euclidean')
model_KNN.fit(X_train,Y_train)
Y_pred=model_KNN.predict(X_test)
print(list(zip(Y_test,Y_pred)))
```

```python
np.sqrt(len(X_train))
```

```python
# Metrics Evaluation

from sklearn.metrics import confusion_matrix,accuracy_score,classification_report

cfm=confusion_matrix(Y_test,Y_pred)
print(cfm)

print()

print("classification report:")
print(classification_report(Y_test,Y_pred))

print()

acc=accuracy_score(Y_test, Y_pred)
print("Accuracy of the model: ",acc)
```

```python
#k Value- Trial and error method

from sklearn.metrics import accuracy_score
my_dict={}
for K in range(1,26):
    model_KNN = KNeighborsClassifier(n_neighbors=K,metric="euclidean")
    model_KNN.fit(X_train, Y_train)
    Y_pred = model_KNN.predict(X_test)
    print ("Accuracy is ", accuracy_score(Y_test,Y_pred), "for K-Value:",K)
    my_dict[K]=accuracy_score(Y_test,Y_pred)
```

```python
for k in my_dict:
    if my_dict[k]==max(my_dict.values()):
        print(k,":",my_dict[k])
```

```python
from sklearn.neighbors import KNeighborsClassifier
model_KNN=KNeighborsClassifier(n_neighbors=9,metric='euclidean')
model_KNN.fit(X_train,Y_train)
Y_pred=model_KNN.predict(X_test)
print(list(zip(Y_test,Y_pred)))
```

```python
# Metrics Evaluation

from sklearn.metrics import confusion_matrix,accuracy_score,classification_report

cfm=confusion_matrix(Y_test,Y_pred)
print(cfm)

print()

print("classification report:")
print(classification_report(Y_test,Y_pred))

print()

acc=accuracy_score(Y_test, Y_pred)
print("Accuracy of the model: ",acc)
```

### Logistic Regression as its binary classification. To Compare which Algorithm is better

```python
from sklearn.linear_model import LogisticRegression
classifier=LogisticRegression()
classifier.fit(X_train,Y_train)
Y_pred=classifier.predict(X_test)
print(list(zip(Y_test,Y_pred)))
```

```python
# Metrics Evaluation

from sklearn.metrics import confusion_matrix,accuracy_score,classification_report

cfm=confusion_matrix(Y_test,Y_pred)
print(cfm)

print()

print("classification report:")
print(classification_report(Y_test,Y_pred))

print()

acc=accuracy_score(Y_test, Y_pred)
print("Accuracy of the model: ",acc)
```
